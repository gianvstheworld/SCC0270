````python
def load_data(df, look_back, test_size):
    '''
    Função que cria os conjuntos de treino e teste e define o comprimento da sequência de análise.
    O comprimento da sequência é o período de tempo (em dias) da previsão.
    Parâmetros:
    df: Pandas DataFrame
    look_back: comprimento da sequência (horizonte de previsão)
    test_size: percentual do conjunto de dados total destinado ao conjunto de teste (0 < test_size < 1).
    '''

    data_raw = df.values    # extrai os valores somente
    data = []               # lista para acrescentar dados

    # cria todas as sequências possíveis de comprimento (horizonte de previsão)
    for index in range(len(data_raw) - look_back): 
        data.append(data_raw[index: index + look_back])

    data = np.array(data);  # converte para numpy array

    # criação dos conjuntos de teste e treino
    test_set_size = int(np.round(test_size*data.shape[0]));
    train_set_size = data.shape[0] - (test_set_size);

    x_train = data[:train_set_size,:-1,:] # define um comprimento de 59 preços de ações
    y_train = data[:train_set_size,-1,:]  # e o 60º é a varível resposta

    x_test = data[train_set_size:,:-1]
    y_test = data[train_set_size:,-1,:]

    # converte para pytorch tensors
    x_train = torch.from_numpy(x_train).type(torch.Tensor)
    x_test = torch.from_numpy(x_test).type(torch.Tensor)
    y_train = torch.from_numpy(y_train).type(torch.Tensor)
    y_test = torch.from_numpy(y_test).type(torch.Tensor)

    return [x_train, y_train, x_test, y_test]
````

````python
look_back = 60 # horizonte de previsão
x_train, y_train, x_test, y_test = load_data(df_aapl, look_back, 0.2)
````

## Modelagem

````python
# defina a seed para garantir a reproducibilidade
torch.manual_seed(3407)
````

Estrutura do modelo LSTM Base

````python
class LSTM_base(nn.Module):

    def __init__(self, input_dim, hidden_dim, num_layers, output_dim):
        super(LSTM_base, self).__init__()
        self.hidden_dim = hidden_dim                                             # dimensões ocultas
        self.num_layers = num_layers                                             # número de camadas ocultas
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True) # batch_first=True faz com que os tensores tenham shape = (batch_dim, seq_dim, feature_dim)
        self.fc = nn.Linear(hidden_dim, output_dim)                              # camada Fully Connected

    def forward(self, x):
        # inicialização com zeros
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim).requires_grad_()
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim).requires_grad_()

        # Precisamos fazer o detach, já que realizamos o backpropagation through time (BPTT) truncado.
        # Se não fizermos isso, o backpropagation vai até o início da série
        out, (hn, cn) = self.lstm(x, (h0.detach(), c0.detach()))

        # Só precisamos ficar com o hidden state do último passo:
        out = self.fc(out[:, -1, :])

        return out
````

````python
# Parametros

input_dim = 1
hidden_dim = 32
num_layers = 2
output_dim = 1
lr = 0.1
num_epochs = 50
````

````python
# Instancia o modelo
model = LSTM_base(input_dim=input_dim, hidden_dim=hidden_dim, output_dim=output_dim, num_layers=num_layers)
````

````python
print('Modelo:\n', model)

print('Quantidade de parâmetros: ', len(list(model.parameters())))
for i in range(len(list(model.parameters()))):
    print(list(model.parameters())[i].size())
````

````python
# Define função de perda
loss_fn = torch.nn.MSELoss()

# Define otimizador
optimizer = torch.optim.SGD(model.parameters(), lr=lr)
````

### **Agora complemente a célula abaixo para implementar uma função que realize o treinamento e faça as previsões**

````python
def train_model(model, num_epochs, x_train, x_test):

    # vetor para registrar as perdas ao longo das épocas
    hist = np.zeros(num_epochs)

    for t in range(num_epochs):
        # implemente o passo forward
        y_train_pred = model(x_train)

        # compute a perda
        loss = loss_fn(y_train_pred, y_train)
        
        # print de algumas épocas
        if t % 10 == 0 and t !=0:
            print("Epoch ", t, "MSE: ", loss.item())

        # armazena a perda
        hist[t] = loss.item()

        # limpe o gradiente
        optimizer.zero_grad()

        # implemente o passo backward
        loss.backward()

        # atualize os parâmetros
        optimizer.step()

    # faça as previsões
    y_test_pred = model(x_test)

    # plot da perda durante o treinamento
    plt.plot(hist, label="Training loss")
    plt.legend()
    plt.show()

    return y_train_pred, y_test_pred, hist
````

````python
y_train_pred, y_test_pred, hist = train_model(model, num_epochs, x_train, x_test)
````

````python
def model_performance(y_train_pred, y_test_pred):
    # aplique a transformação reversa para os dados voltarem à sua escala original
    y_train_rev = scaler.inverse_transform(y_train.detach().numpy())
    y_test_rev = scaler.inverse_transform(y_test.detach().numpy())
    y_train_pred_rev = scaler.inverse_transform(y_train_pred.detach().numpy())
    y_test_pred_rev = scaler.inverse_transform(y_test_pred.detach().numpy())

    # calcule o RSME para os conjuntos de treino e teste
    trainScore = math.sqrt(mean_squared_error(y_train_rev, y_train_pred_rev))
    print('Train Score: %.2f RMSE' % (trainScore))
    testScore = math.sqrt(mean_squared_error(y_test_rev, y_test_pred_rev))
    print('Test Score: %.2f RMSE' % (testScore))

    # Plot dos resultados
    figure, axes = plt.subplots(figsize=(15, 6))
    axes.xaxis_date()

    axes.plot(df_aapl[len(df_aapl)-len(y_test):].index, y_test_rev, color = 'red', label = 'Real Apple Stock Price')
    axes.plot(df_aapl[len(df_aapl)-len(y_test):].index, y_test_pred_rev, color = 'blue', label = 'Predicted Apple Stock Price')
    plt.title('Apple Stock Price Prediction')
    plt.xlabel('Time')
    plt.ylabel('Apple Stock Price')
    plt.legend()
    plt.show()
````

````python
# execute essa célula
model_performance(y_train_pred, y_test_pred)
````

### Hiperparâmetros relevantes para tunar em uma LSTM
Fonte: https://medium.com/geekculture/10-hyperparameters-to-keep-an-eye-on-for-your-lstm-model-and-other-tips-f0ff5b63fcd4

- Número de dimensões ocultas: como regra geral, uma camada oculta irá funcionar para os problemas mais simples e duas camadas para problemas com complexidade razoável;

- Dropout: toda camada LSTM deve ser acompanhada por uma camada de dropout para evitar overfitting. O dropout é uma técnica de regularização que consiste em zerar aleatoriamente algumas entradas da camada anterior; 

- Taxa de aprendizado: esse hiperparâmetro é o mais importante para o treinamento de uma rede neural. A taxa de aprendizado é o tamanho do passo que o algoritmo de otimização irá dar para encontrar o mínimo da função de custo. Se a taxa de aprendizado for muito pequena, o algoritmo irá convergir lentamente. Se a taxa de aprendizado for muito alta, o algoritmo irá divergir;

- Momentum: é o único hiperparâmetro que permite o acumúlo de gradientes ao longo do tempo. O momentum é uma técnica de otimização que consiste em adicionar uma fração do gradiente anterior ao gradiente atual. Isso permite que o algoritmo de otimização se mova mais rápido em direção ao mínimo da função de custo. 

````python
class LSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, output_dim):
        super(LSTM, self).__init__() 
        self.hidden_dim = hidden_dim # dimensões ocultas
        self.num_layers = num_layers # número de camadas
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True) # camada da Long Short-Term Memory
        self.fc = nn.Linear(hidden_dim, output_dim) # camada Fully Connected
        self.dropout = nn.Dropout(0.2) # 20% de dropout

    def forward(self, x):
        # inicialização com zeros
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim).requires_grad_()
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim).requires_grad_()

        # Precisamos fazer o detach, já que realizamos o backpropagation through time (BPTT) truncado.
        # Se não fizermos isso, o backpropagation vai até o início da série
        out, (hn, cn) = self.lstm(x, (h0.detach(), c0.detach()))

        # Só precisamos ficar com o hidden state do último passo:
        out = self.dropout(out[:, -1, :])
        out = self.fc(out)

        return out
````

### Alterações feitas em relação ao modelo base

- **Número de épocas**: 50 -> 100 - O número de épocas foi aumentado para que o modelo pudesse aprender melhor os padrões dos dados. Além disso, é interessante comentar que o número de épocas não é um hiperparâmetro que deve ser definido de forma fixa. O número de épocas deve ser definido de forma que o modelo não apresente overfitting, ou seja, que o modelo não apresente um erro de treinamento muito baixo e um erro de validação muito alto.
- **Taxa de aprendizado**: 0.1 -> 0.01625 
- **Otimizador**: SGD -> Adam - Para lidar melhor com a complexidade dinâmica do treinamento de RNN, otimizadores que utilizam o momentum e possuem taxa de aprendizado adaptativa como o Adam são melhores do que o SGD.
